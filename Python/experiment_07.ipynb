````python
from my_package.Account.salary import Salary
from my_package.Employee.qualification import Qualification
from my_package.Employee.profile import Profile

n = int(input("Enter number of Employees: "))
emp_list = []
for i in range(n):
    
    name = input(f"Enter Name of Employee {i+1}: ")
    age = input(f"Enter Age of Employee {i+1}: ")
    dob = input(f"Enter Birth Year of Employee {i+1}: ")
    degree = input(f"Enter Degree of Employee {i+1}: ")
    experience = input(f"Enter Experience of Employee {i+1} (in years): ")
    basic = int(input(f"Enter Basic Pay of Employee {i+1}: "))
    hra = int(input(f"Enter HRA of Employee {i+1}: "))
    pf = int(input(f"Enter PF of Employee {i+1}: "))
    
    emp_prof = Profile(name, age, dob)
    emp_q = Qualification(degree, experience)
    emp_s = Salary(pf, basic, hra)
    
    t_salary = basic + hra - pf
    
    new_employee = [emp_prof, emp_q, emp_s, t_salary]
    emp_list.append(new_employee)
    print()

print("Name\tAge\tDob\tDegree\tExp\tBasic\tHRA\tPF\tSalary")
print('-'*75)
for emp in emp_list:
    print(emp[0].name, emp[0].age, emp[0].dob, emp[1].degree, emp[1].qualification, emp[2].basic_pay, emp[2].HRA, emp[2].PF, emp[3], sep='\t')
````

## Question 2

Python program to demonstrate use of regular expression 

- Create string with name of cities in india separated by spaces. 
- Find all cities ending with “ai”
- Find all cities starting with “Mu” or “Ma”
- print name of cities with ‘u’ as second letter and ‘a’ as second last letter 

````python
import re
city = "Mumbai Surat Pune Delhi Chennai Indore Hyderabad Jaipur Kanpur Banglore Ranchi Amritsar Kolkata Manglore Mathura"
cities = city.split()

print("Cities Ending with 'ai':")
print('-'*30)
for city in cities:
    check = re.search('ai$', city)
    if check:
        print(city)
        
print("\nCities starting with 'Mu' or 'Ma':")
print('-'*40)
for city in cities:
    check = re.search('(^mu)|(^ma)', city.lower())
    if check:
        print(city)
        
print("\nCities with 'u' as second letter and 'a' as second last:") 
print('-'*60)
for city in cities:
    check = re.search('(^.u).*(a.$)', city.lower())
    if check:
        print(city)
````

## Question 3

Python program to demonstrate use of regular expression 

- Create a phone list using file (surname    name    number)
- Find  all  the  entries  of  phone  book  with  surname  as “Rao”and  first  name starting with ‘J’ or ‘K’

````python
phone_list = []
with open('PhoneBook.txt', 'r') as my_file:
    
    for entry in my_file:
        phone_list.append(entry.split())

print(phone_list)

print('\nSurname\tName\tPhone')
print('-'*30)

for phone in phone_list:
    
    check = re.search('(^J.)|(^K.)', phone[1]) 
    if check and phone[0] == 'Rao':
        
        print(f'{phone[0]}\t{phone[1]}\t{phone[2]}')
````
