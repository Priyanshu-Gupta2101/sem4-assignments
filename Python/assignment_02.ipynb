## Question 2

Menu driven code for numpy array

● create using array() and arange()

● sum of array 

● sort array

● compare two arrays

```python
import numpy as np

def menu():
    print("\n[1]. Create a NumPy Array")
    print("[2]. Sum of the NumPy Array")
    print("[3]. Sort the NumPy Array")
    print("[4]. Compare NumPy Arrays")
    print("[5]. Exit")
    choice = int(input("Enter Choice: "))
    return choice

print("Welcome To NumPy Operations")
while(True):
    ar = np.array([])
    ch = menu()
    if(ch == 1):
        choi = int(input("Create an Array \n[1]. Using the arange function \n[2]. Enter the array manually\nEnter Choice:"))
        if choi == 1:
            n = int(input("Enter a number: "))
            arr = np.arange(n)
            print(f"Numpy Array created: {arr}")
        else:
            n = int(input("Enter a number: "))
            arr = np.array([int(x) for x in input("Enter n elements: ").split()])
            print(f"Numpy Array created: {arr}")
        
    elif(ch == 2):
        print(f"The sum of the array is: {arr.sum()}")
        
    elif(ch == 3):
        arr.sort()
        print(f"Array before sorting: {arr}")
        arr.sort()
        print(f"Array before sorting: {arr}")
        
    elif(ch == 4):
        a = np.array([101, 99, 87])
        b = np.array([897, 97, 111])
        
        print("Array a: ", a)
        print("Array b: ", b)
        
        print("a > b")
        print(np.greater(a, b))

        print("a >= b")
        print(np.greater_equal(a, b))

        print("a < b")
        print(np.less(a, b))

        print("a <= b")
        print(np.less_equal(a, b))
    
    else:
        print("Exiting...")
        break
        
```

## Question 3

Python program

● To read a two matrices from user 

● Perform matrix multiplication 

● Display diagonal Elements 

● Check whether its a square matrix

```python
import numpy as np

while True:
    choice = int(input("""
1. Enter two matrices.
2. Matrix multiplication..
3. Display diagonal elements.
4. Check square matrix.
Enter Choice: """))

    if choice == 1:
        print("\nFor First Matrix")
        R = int(input("Enter the number of rows: "))
        C = int(input("Enter the number of columns: "))

        li = [int(x) for x in input("Enter the entries in a single line (separated by space): ").split()]
        
        print("\nFor Second Matrix ")
        R2 = int(input("Enter the number of rows: "))
        C2 = int(input("Enter the number of columns: "))

        li2 = [int(x) for x in input("Enter the entries in a single line (separated by space): ").split()]
        
        matrix = np.array(li).reshape(R, C)
        matrix2 = np.array(li2).reshape(R2, C2)
    
        print("Matrices are: ")
        print(matrix)
        print(matrix2)

    elif choice == 2:
        print("Product of matrices is: ")
        print(np.dot(matrix, matrix2))

    elif choice == 3:
        print("Diagonal of matrix1 is: ", np.diag(matrix))
        print("Diagonal of matrix2 is: ", np.diag(matrix2))
        
    elif choice == 4:
        if R == C:
            print("Matrix1 is square.")
        else:
            print("Matrix1 is not a square.")

        if R2 == C2:
            print("Matrix2 is square.")
        else:
            print("Matrix2 is not a square.")
    else:
        print("Exiting...")
        break
```
