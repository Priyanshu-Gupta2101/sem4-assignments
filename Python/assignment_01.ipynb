## Question 1
Python program to implement operations of linked list 
1. Display list
2. Insert at beginning 
3. Insert at End 
4. Insert at specified postion 
5. Delete from beginning 
6. Delete from end 
7. Delete at specified postion 
8. Delete a particular element 
9. Search an element 
10. Replace element at specified index  
11. Forward traversal
12. Reverse traversal 
13. Insert after an element

```python
class LinkedList:
    
    # Constructor
    def __init__(self):
        self.linked_list = []
        
    # Helper Methods
    def print_linked_list(self):
        print("\nLinked List: ", end = "")
        
        if self.linked_list_empty():
            print("Empty")
            return
        
        for i in (self.linked_list):
            print(f" {i} -> ", end = "")
        print("NONE")
        
    def print_reverse_linked_list(self):
        print("\nLinked List: ", end = "")
        
        if self.linked_list_empty():
            print("Empty")
            return
        
        for i in reversed(self.linked_list):
            print(f" {i} <- ", end = "")
        
    def print_size_of_linked_list(self):
        print(f"\nSize of LinkedList is {self.size_of_linked_list()}")
        
    def find_in_linked_list(self, value):
        pos = self.search_linked_list(value)
        if pos == -1:
            print(f"The element {value} does not exists in the Linked List.")
        else:
            print(f"The value {value} was found at {pos + 1}th position in the Linked List.")
            
        self.print_linked_list()
        
    
    # Setter methods
    def append_end(self, element):
        print(f"\nAdding element {element} at the end in the Linked List...")
        self.linked_list.append(element)
        
        self.print_linked_list()
        
    def append_beginning(self, element):
        print(f"\nAdding element {element} at the beginning in the Linked List...")
        self.linked_list.insert(0, element)
        
        self.print_linked_list()
        
    def append_at_position(self, element, position):
        print(f"\nAdding element {element} at the position {position} in the Linked List...")
        self.linked_list.insert(position - 1, element)
        
        self.print_linked_list()

        
        
    def remove_end(self):
        if not self.linked_list_empty():
            print(f"\nRemoving the last element {self.linked_list[-1]} from the Linked List...")
            self.linked_list.pop()
            
        self.print_linked_list()
        
    def remove_beginning(self):
        if not self.linked_list_empty():
            print(f"\nRemoving the first element {self.linked_list[0]} from the Linked List...")
            del self.linked_list[0]
            
        self.print_linked_list()
        
    def remove_at_position(self, position):
        if not self.linked_list_empty():
            if position - 1 < self.size_of_linked_list():
                print(f"\nRemoving the {position} element {self.linked_list[position - 1]} from the Linked List...")
                del self.linked_list[position - 1]
            else:
                print("\nInvalid Position.")
            
        self.print_linked_list()
        
    def remove_specific(self, element):
        if not self.linked_list_empty():
            if element in self.linked_list:
                print(f"\nRemoving the element {element} from the Linked List...")
                del self.linked_list[0]
            else:
                print("\nElement not present.")
            
        self.print_linked_list()
        
    def replace_element_at_index(self, index, value):
        if index - 1 < self.size_of_linked_list():
            print(f"\nReplacing the element {self.linked_list[index - 1]} at {index - 1} by {value} in the Linked List...")
            self.linked_list[index - 1] = value
        else:
            print("\nInvalid Position.")
        
    
    # Getter methods
    def size_of_linked_list(self):
        return len(self.linked_list)
    
    def search_linked_list(self, value):
        j = -1
        if value not in self.linked_list:
            return j
        for i in (self.linked_list):
            if i == value:
                return j + 1
            j += 1
    
    def linked_list_empty(self):
        return self.size_of_linked_list() == 0
```

```python
def menu():
    print("Welcome to LinkedList Operations")
    
    # Initialize Empty LinkedList
    print("\nCreating an Instance of a LinkedList...")
    instance = LinkedList()
    print("Empty Linked List Initialized")
    
    while(True):
        
        # Menu
        print("\nLinkedList Operations:")
        print("[1]  Print LinkedList")
        print("[2]  Insert at beginning")
        print("[3]  Insert at End")
        print("[4]  Insert at specified postion")
        print("[5]  Delete from beginning")
        print("[6]  Delete from End")
        print("[7]  Delete at specified position")
        print("[8]  Delete a particular element")
        print("[9]  Search an element")
        print("[10] Replace element at specified index")        
        print("[11] Print reverse LinkedList")
        print("[12] Destroy LinkedList")
        print("[13] Exit")
        
        choice = int(input("Enter a number [1-13]: "))
        if choice == 1:
            instance.print_linked_list()
            
        elif choice == 2:
            value = int(input("Enter a number: "))
            instance.append_beginning(value)
        elif choice == 3:
            value = int(input("Enter a number: "))
            instance.append_end(value)
        elif choice == 4:
            value = int(input("Enter a number: "))
            position = int(input("Enter an index: "))
            instance.append_at_position(value, position)
            
        elif choice == 5:
            instance.remove_beginning()
        elif choice == 6:
            instance.remove_end()
        elif choice == 7:
            position = int(input("Enter an index: "))
            instance.remove_at_position(position)
        elif choice == 8:
            value = int(input("Enter an element: "))
            instance.remove_specific(value)
            
        elif choice == 9:
            search = int(input("Enter an element to be searched: "))
            instance.find_in_linked_list(search)
        
        elif choice == 10:
            value = int(input("Enter a number: "))
            position = int(input("Enter an index: "))
            instance.replace_element_at_index(position, value)
            
        elif choice == 11:
            instance.print_reverse_linked_list()
            
        elif choice == 12:
            print("\nDestroying LinkedList...")
            instance = LinkedList()
            print("\nLinkedList destroyed. New instance created.")
            
        elif choice == 13:
            return
        
        else:
            print("Enter a number between [1-13].")
            
menu()
```

## Question 2

Python program to implement `Stack` data structure, create `class Stack` with following functions.
- Push an element
- Pop an element
- Top of Stack
- Search an element

Display stack create instance and perform all operations.

```python
class Stack:
    
    # Constructor
    def __init__(self):
        self.stack = []
        
    # Helper Methods
    def print_stack(self):
        print("\nCurrent Stack: ")
        if self.stack_empty():
            print("Stack Empty")
            return
        
        for i in reversed(self.stack):
            print(f"| {i} |")
        print("-----")
        
    def print_top_of_stack(self):
        if not self.stack_empty():
            print(f"Element at top of the stack is {self.top_of_stack()}")
        else:
            self.print_stack()
        
    def print_size_of_stack(self):
        print(f"Element at top of the stack is {self.size_of_stack()}")
        
    def find_in_stack(self, value):
        pos = self.search_stack(value)
        if pos == -1:
            print(f"The element {value} does not exists in the Stack.")
        else:
            print(f"The value {value} was found at {pos + 1}th position from the top of the stack.")
            
        self.print_stack()
        
    
    # Setter methods
    def push_element(self, element):
        print(f"\nPushing element {element} on the stack...")
        self.stack.append(element)
        self.print_stack()
    
    def pop_element(self):
        if not self.stack_empty():
            print(f"\nPopping element {self.top_of_stack()} from the stack...")
            self.stack.pop()
            
        self.print_stack()
        
    
    # Getter methods
    def size_of_stack(self):
        return len(self.stack)
    
    def top_of_stack(self):
        return self.stack[-1]
    
    def search_stack(self, value):
        j = -1
        if value not in self.stack:
            return j
        for i in reversed(self.stack):
            if i == value:
                return j + 1
            j += 1
    
    def stack_empty(self):
        return self.size_of_stack() == 0
```

```python
def menu():
    print("Welcome to Stack Operations")
    
    # Initialize Empty Stack
    print("\nCreating an Instance of a stack...")
    instance = Stack()
    print("Empty Stack Initialized")
    
    while(True):
        
        # Menu
        print("\nStack Operations:")
        print("[1] Print Stack")
        print("[2] Push Element on the Stack")
        print("[3] Pop Element off the Stack")
        print("[4] Top of the Stack")
        print("[5] Find Element in the Stack")
        print("[6] Destroy Stack")
        print("[7] Exit")
        
        choice = int(input("Enter a number [1-7]: "))
        if choice == 1:
            instance.print_stack()
        elif choice == 2:
            value = int(input("Enter a number: "))
            instance.push_element(value)
        elif choice == 3:
            instance.pop_element()
        elif choice == 4:
            instance.print_top_of_stack()
        elif choice == 5:
            search = int(input("Enter an element to be searched: "))
            instance.find_in_stack(search)
        elif choice == 6:
            print("\nDestroying Stack...")
            instance = Stack()
            print("\nStack destroyed. New instance created.")
        elif choice == 7:
            return
        else:
            print("Enter a number between [1-7].")
            
menu()
```

## Question 3

Python program to implement `Queue` data structure, create `class Queue` with following functions.
- Insert an element
- Remove an element
- Display Queue
- Search an element

Display stack create instance and perform all operations.

```python
class Queue:
    
    # Constructor
    def __init__(self):
        self.queue = []
        
    # Helper Methods
    def print_queue(self):
        print("\nCurrent Queue: ", end = "")
        if self.queue_empty():
            print("Empty")
            return
        print("|", end = "")
        for i in (self.queue):
            print(f" {i} |", end = "")
        
    def print_front_of_queue(self):
        if not self.queue_empty():
            print(f"Element at front of the queue is {self.front_of_queue()}")
        else:
            self.print_queue()
        
    def print_size_of_stack(self):
        print(f"Size of the queue is {self.size_of_stack()}")
        
    def find_in_queue(self, value):
        pos = self.search_queue(value)
        if pos == -1:
            print(f"The element {value} does not exists in the Queue.")
        else:
            print(f"The value {value} was found at {pos + 1}th position from the front of the Queue.")
            
        self.print_queue()
        
    
    # Setter methods
    def add_element(self, element):
        print(f"\nAdding element {element} in the Queue...")
        self.queue.append(element)
        self.print_queue()
    
    def remove_element(self):
        if not self.queue_empty():
            print(f"\nRemoving element {self.front_of_queue()} from the Queue...")
            del self.queue[0]
            
        self.print_queue()
        
    
    # Getter methods
    def size_of_queue(self):
        return len(self.queue)
    
    def front_of_queue(self):
        return self.queue[0]
    
    def search_queue(self, value):
        j = -1
        if value not in self.queue:
            return j
        for i in (self.queue):
            if i == value:
                return j + 1
            j += 1
    
    def queue_empty(self):
        return self.size_of_queue() == 0
```

```python
def menu():
    print("Welcome to Queue Operations")
    
    # Initialize Empty Queue
    print("\nCreating an Instance of a Queue...")
    instance = Queue()
    print("Empty Queue Initialized")
    
    while(True):
        
        # Menu
        print("\nQueue Operations:")
        print("[1] Print Queue")
        print("[2] Add Element")
        print("[3] Remove Element")
        print("[4] Display Front of the Queue ")
        print("[5] Find Element in the Queue")
        print("[6] Destroy Queue")
        print("[7] Exit")
        
        choice = int(input("Enter a number [1-7]: "))
        if choice == 1:
            instance.print_queue()
        elif choice == 2:
            value = int(input("Enter a number: "))
            instance.add_element(value)
        elif choice == 3:
            instance.remove_element()
        elif choice == 4:
            instance.print_front_of_queue()
        elif choice == 5:
            search = int(input("Enter an element to be searched: "))
            instance.find_in_queue(search)
        elif choice == 6:
            print("\nDestroying Queue...")
            instance = Queue()
            print("\nQueue destroyed. New instance created.")
        elif choice == 7:
            return
        else:
            print("Enter a number between [1-7].")
            
menu()
```

# Question 4
Python program to use deque class from collections with following functions.
- Add element at Front
- Remove element from Front
- Add element at Rear
- Remove element from Rear
- Search for an element

```python
from collections import deque
d = deque()
choice = 0
while choice<6:
    
    print('\nDEQUE OPERATIONS')
    print('[1] Insert an element at front')
    print('[2] Remove an element from front')
    print('[3] Insert an element at rear')
    print('[4] Remove an element from rear')
    print('[5] Search an element')
    print('[6] Exit')
    choice = int(input('Your choice: '))
    if choice==1:
        element = int(input('Enter element: '))
        d.appendleft(element)
        
    elif choice==2:
        if len(d) == 0:
            print('Deque is empty')
        else:
            d.popleft()
            
    elif choice==3:
        element = int(input('Enter element: '))
        d.append(element)
        
    elif choice==4:
        if len(d) == 0:
            print('Deque is empty')
        else:
            d.pop()
            
    elif choice==5:
        element = int(input('Enter element: '))
        c = d.count(element)
        print('No of times the element found:', c)
    else:
        break
    
    print('Deque: ' , end='')
    for i in d:
        print(i, ' ', end='')
    print()
```
